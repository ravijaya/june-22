```python
import pandas as pd

fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
sectors.size()
```

```python
fortune['Sector'].value_counts()
```

```python
sectors.groups
```

```python
sectors.get_group('Energy')  # to get an group info
sectors.get_group('Chemicals')  # to get an group info
```

# iterating the group by object

   

```python
df = pd.DataFrame(columns=fortune.columns)

for group, data in  sectors:
    highest_revenue_in_group = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue_in_group)
    
df.sort_values('Revenue', ascending=False)
```

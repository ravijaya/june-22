```python
customers = sales[['name','ext price','date']]
customers
```

```python
customer_group = customers.groupby('name')
customer_group
```

```python
sales_totals = customer_group.sum()
sales_totals
```

```python
op = sales_totals.sort_values(by='ext price', ascending=False)
op
```

```python
bar_plot = op.plot(kind='bar',legend=None, title="Total Sales by Customer")

bar_plot.set_xlabel("Customers")
bar_plot.set_ylabel("Sales ($)")
#plt.show()
plt.savefig('cust.png')
```
